```python
import pandas as pd 
import joblib
import matplotlib.pyplot as plt
import os
```

```python
df_fps = joblib.load('../data/raw/CSD_melting_point/melting_point_fps.job')
df_rdkit_2d = joblib.load('../data/raw/CSD_melting_point/melting_point_rdkit_2d.job')
```

These two files are binary files containing our datasets. They are as follows (according to the github of the authors): 

1. df_fps -- a binary file containing RDKit radius 2, 1024 bits fingerprint
2. df_rdkit_2d -- another binary file containing RDKit 2D descriptors (209)

Now let's investigate the type of these files and how many data points they both contain. 

```python
print(type(df_fps))
print(df_fps.shape)
```

```python
print(type(df_rdkit_2d))
print(df_rdkit_2d.shape)
```

Let's investigate both of these dataframes and what data they actually contain. 

## first dataframe

```python
df_fps.head()
```

```python
df_fps.info()
```

This dataframe contains 588180 rows, one for each molecule, with the index containing the CID# pertaining to each molecule. There are four rows, each containing a piece of information about the molecule. 

1. std_temp -- standardized melting point (mp) in Kelvin 
2. FP -- the morgan fingerprints of each molecule, radius 2, 1024 bits 
3. clustering -- whether the molecule belongs to the training (TRN) or test (TST) set based on the clustering method used by the authors 
4. random -- whether the molecule belongs to the training (TRN) or test (TST) set based on random division 

## second dataframe

```python
df_rdkit_2d.head()
```

```python
df_rdkit_2d.info()
```

The second dataframe also contains 588180 entries, one for each molecule in our dataset, with the index corresponding to the CID# pertaining to each unique molecule. There are 211 rows, described as follows: 
1. 209 RDKit molecular descriptors (2D) -- these include descriptors such as the number of valence electrons, the molecular weight, the number of heavy atoms in the molecule, the aromaticity of the molecule, etc. (definitions for these can be found here: https://www.rdkit.org/docs/source/rdkit.Chem.rdMolDescriptors.html)
2. std_temp -- standardized melting point (mp) in Kelvin
3. clustering -- whether the molecule belongs to the training (TRN) or test (TST) set based on the clustering method used by the authors 
4. random -- whether the molecule belongs to the training (TRN) or test (TST) set based on the random method used by the authors 

Note from the printed information about these dataframes, neither contain any missing values, so we will not have to deal with those. 

The two dataframes have four features in common:
1. the CID# of each molecule
2. the std_temp of each molecule in K
3. clustering assignment
4. random assignment 

It will be worth to check whether or not these values match for each of the CID#'s, as a quick sanity check for the validity of this data. 

## Joining the dataframes

First, let's run a check that the values for the std_temp, clustering assignment, and random assignment are the same for each CID # between the two dataframes. We can do this by performing a left join on the first dataframe on index, std_temp, clustering, and random, and then checking to make sure all 588180 rows made the final, joined dataframe. 

First, let's make the index it's own columm. 

```python
df_fps = df_fps.reset_index().rename(columns={"index":"CID"})
df_fps.head()
```

```python
df_rdkit_2d = df_rdkit_2d.reset_index().rename(columns={"index":"CID"})
df_rdkit_2d.head()
```

Now to join the two dataframes. 

```python
df_fps_rdkit_2d = pd.merge(df_fps, df_rdkit_2d, how='left', on=['CID', 'std_temp', 'clustering', 'random'])
df_fps_rdkit_2d.head()
```

```python
df_fps_rdkit_2d.info()
```

Since we have kept all 588180 entries, we know that all the data matched, so the two tables matched and were successfully joined.

Finally, we will remove the columns provided by the authors giving the test and training sets they assigned, as we will attempt to assign our own test/training sets. 

```python
df_fps_rdkit_2d.drop(columns=['clustering', 'random'], inplace=True)
df_fps_rdkit_2d.info()
```

```python
datasets = os.listdir('../data/raw/external')
print(datasets)
```

```python
print(len('SI_dataset_'))
```

```python
dfs = []
for dataset in datasets:
    df = pd.read_csv(f'../data/raw/external/{dataset}')
    print(df.columns)
    print(df.head())
    
    column_names = list(df.columns)
    for col in column_names:
        if col == 'Structure':
            index = column_names.index('Structure')
            column_names[index] = 'SMILES'
        elif col == 'ID':
            index2 = column_names.index('ID')
            column_names[index2] = 'Name'
    df.columns = column_names
    
    df2 = df[['SMILES', 'Name', 'logS_expr', 'LogP_ACD', 'SFI']]
    idx = dataset[11:-4]
    df2['dataset'] = [idx for i in range(len(df2))]
    dfs.append(df2)
```

```python
val = pd.concat(dfs)
val.head()
```

```python
val.info()
```

```python
val.dataset.value_counts()
```

```python
val.to_csv('../data/intermediate/external_validation_datasets.csv')
```

## Summary 

The original two dataframes imported were one dataframe containing the morgan fingerprints for each molecule (radius 2, 1024 bits) and another containing 209 2D molecular descriptors from RDKit for each molecule. There are 588180 molecules total in the dataset, each with its own CID #, std_temp (which corresponds to the standardized melting point in Kelvin), and a designation for whether the creators of the dataset would assign it to a training or test set. 

First, the two dataframes were joined on the unique CID #'s of the molecules, as well as their std_temp and training/test assignment. Next, we identified our target feature, the std_temp, and looked at its distribution, which appears to be normal. 

Next, we looked at the distributions of all the other features contained within our dataset, particularly the molecular descriptors. Although we identified many features which contain a majority of values within a certain range and a smaller number of outliers, we did not choose to eliminate any of those features at this time. However, we did identify three molecular descriptors for which all values in all the rows were equal to zero, and we did drop those three columns. We also dropped the two columns indicating whether the creators of the dataset would choose to assign the molecules to the test/training datasets, as we will want to investigate these features and decide our own training/test split moving forward. 

The final dataframe saved contains 588180 entries for each of the molecules, containing 206 molecular descriptors and the morgan fingerprints for each molecule, as well as the std_temp, our feature of interest. There is enough data to make a model for this achievable.
